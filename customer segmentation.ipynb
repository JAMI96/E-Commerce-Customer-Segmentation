```python
x=df.iloc[:,2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

```python
#how to find out the k value
inertia=[]
for i in range(1,16):
  k_means = KMeans(n_clusters=i)
  k_means = k_means.fit(features)
  inertia.append(k_means.inertia_)
```

```python
#Elbow Graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.xlabel('No of clusters'),plt.ylabel('Inertia')
```

```python
#kelbow visualizer
plt.subplot(1,2,2)
kmeans=KMeans()
visualize=KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
```

```python
silhoutte_avg=[]
for i in range(2,16):
  k_means=KMeans(n_clusters=i)
  cluster_labels=k_means.fit_predict(features)
#silhouette score
  silhoutte_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhoutte_avg,'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhoutte Score')
plt.title('Silhoutte Analysis for optimal K')
plt.show()
```

Kmeans model where k value will be 3 as per elbow method

```python
model =KMeans(n_clusters=3)
model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```

```python
df['cluster']=pd.DataFrame(y_km)
df.to_csv('Cluster_data.csv',index=False)
```
